```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
data = load_boston()
boston = pd.DataFrame(data.data)
boston.columns = data.feature_names
boston["Price"]=data.target
```

```python
y =boston.loc[:,'Price'].values
X = boston.iloc[:,:13].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lcr = LinearRegression()
```

```python
lcr.fit(X_train,y_train)
```

```python
prediction = lcr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(y_test,prediction)
```

```python
plt.scatter(y_test,prediction)
plt.show()
```
