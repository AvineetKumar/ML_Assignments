```python
import pandas as pd
```

```python
sales = pd.DataFrame({"TV": [230.1,44.5,17.2,151.5,180.8,8.7,57.5,120.2,8.6,199.8,66.1,214.7,23.8,97.5,204.1],
"Radio": [37.8,39.3,45.9,41.3,10.8,48.9,32.8,19.6,2.1,2.6,5.8,24,35.1,7.6,32.9],
"Newspaper": [69.2,45.1,69.3,58.5,58.4,75,23.5,11.6,1,21.2,24.2,4,65.9,7.2,46],
"Sales": [22.1,10.4,9.3,18.5,12.9,7.2,11.8,13.2,4.8,10.6,8.6,17.4,9.2,9.7,19]})
```

```python
print(sales.TV.corr(sales.Sales))
print(sales.Radio.corr(sales.Sales))
print(sales.Newspaper.corr(sales.Sales))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = sales.iloc[:,:3].values
Y = sales.iloc[:,-1].values
```

```python
lc = LinearRegression()
```

```python
lc.fit(X,Y)
```

Given the following data set, use the sklearn package to find the co-efficients of the
line that describes the following relationships:

1) x1 and y

2) x2 and y

3) x3 and y

```python
lc.coef_
```

```python
pred = lc.predict(X)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(sales, x_vars=['TV','Radio','Newspaper'], y_vars='Sales', height=7, kind='reg')
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(Y,pred)
```

```python
mean_squared_error(Y,pred)
```

```python
sales['Predicted']=pred
```

```python
sns.pairplot(sales, x_vars=['TV','Radio','Newspaper'], y_vars='Predicted', height=7, kind='reg')
plt.show()
```
